```python
month_name = {'01':'Jan', '02':'Feb', '03':'Mar', '04':'Apr', '05':'May', '06':'Jun', '07':'Jul', '08':'Aug', '09':'Sep', '10':'Oct','11':'Nov', '12':'Dec'}

def new(df,Year):
    '''
    Arguments :
      df : pandas dataframe
      years : list containing current_year and previous_year i.e [2021,2022]
    Returns :
      Pandas Dataframe containing data of current year and previous year
    '''
#     dF = df.dropna(subset = ['registration_date'])
    if not np.dtype('datetime64[ns]') == df.order_purchase_timestamp.dtype :
        df['order_purchase_timestamp'] = pd.to_datetime(df.order_purchase_timestamp,errors = 'coerce')
    first = True
    for epoch in Year:
        df1 = df[df.order_purchase_timestamp.dt.year == epoch]
        data = df1.groupby(pd.Grouper(key = 'order_purchase_timestamp',freq = 'M')).agg({'customer_id': pd.Series.nunique})
        data.index = data.index.strftime("%b")
        dt = data.fillna(0)
        dt.index.names = ['month_name']
        if first:
            return_dF = dt.rename({'customer_id':str(epoch)}, axis = 1)
            first = False
        else:
            return_dF = pd.concat([return_dF,dt.rename({'customer_id':str(epoch)}, axis =1)], axis =1)
    return_dF = pd.DataFrame(return_dF, columns = return_dF.columns, index = list(Month_name.values())).replace([np.nan,np.inf,-np.inf],'') 
    return return_dF
```

```python
# Ho : mean of s1 = mean of s2 = mean of s3= mean of s4
# Ha : m1 <> m2 <> m3 <> m4

#function:scipy.stats

#test:f_oneway
    
# CI : 95%, p = 0.05

# p<0.05 to reject null hypothesis
```

```python
credit_card=df.sales[df['payment_type']=='credit_card']
credit_card
```

```python
UPI=df.sales[df['payment_type']=='UPI']
UPI
```

```python
voucher=df.sales[df['payment_type']=='voucher']
voucher
```

```python
debit_card=df.sales[df['payment_type']=='debit_card']
debit_card
```

```python
print(credit_card.mean(), '\n', UPI.mean(), '\n' ,voucher.mean(), '\n', debit_card.mean())
```

```python
import scipy.stats as stats
```

```python
stats.f_oneway(credit_card,UPI,voucher,debit_card)
```

```python
#conclusion: p<0.05, we can reject null hypothesis and accpet alternative hypothesis
#all payment types are not equally used by customers 
#credit card payment method is more widely used by customers
```
